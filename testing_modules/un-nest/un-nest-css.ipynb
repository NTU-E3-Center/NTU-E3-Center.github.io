```python
import re

def unnest_css(input_css):
    """
    Unnests CSS rules from the given CSS string.
    """
    # Split the CSS by lines
    lines = input_css.splitlines()
    
    # Stack to store nested selectors
    selector_stack = []
    
    # Result list to store the unnested CSS
    result_css = []
    
    for line in lines:
        line = line.strip()
        
        if not line:
            continue
        
        if '{' in line:
            # Split the line by '{' to separate selector and the start of the block
            selector, _ = line.split('{', 1)
            selector = selector.strip()
            
            if selector_stack:
                # Combine current selector with the previous one to form the full selector
                full_selector = f'{selector_stack[-1]} {selector}'.strip()
            else:
                full_selector = selector
            
            selector_stack.append(full_selector)
            result_css.append(f'{full_selector} {{')
        
        elif '}' in line:
            selector_stack.pop()
            result_css.append('}')
        
        else:
            # This is a CSS property, just add it to the result
            result_css.append(line)
    
    return '\n'.join(result_css)

def read_css_file(file_path):
    with open(file_path, 'r') as file:
        input_css = file.read()
    return input_css

# Example usage:
file_path = '../../css/style.css'  # Replace with your CSS file path
input_css = read_css_file(file_path)

output_css = unnest_css(input_css)

with open('unnested_styles.css', 'w') as file:
    file.write(output_css)

print("Unnested CSS has been exported to 'unnested_styles.css'")
```
